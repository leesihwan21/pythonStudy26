```python
from idlelib.editor import keynames

# 딕셔너리 타입은 key:value 로 이루어진다.
# { key1 : value1, key2 : value2 ..... }
# 대부분 웹 프로그래밍 언어는 json 이라는 값으로 처리하는데 유사한 형태이다.



dic = {"name" : "kkw", "age" : 40}
#       key       value key    value

type(dic)
print(dic["name"])

#           key
```

```python
a = {1 : 'hi'}
print(a[1])

#      key
```

```python
b = {"a" : [1,2,3]}
#     k       v

print(b["a"])
```

```python
print(b["a"][1])
```

```python
print(b["a"][2])
```

```python
print(dic)
```

```python
# dic 변수에 email key를 갖는 값을 추가하는 방법.
dic['email'] = "mbc@kkw.com"
print(dic)
```

```python
# dic 변수에 age를 삭제.
del dic["age"]
print(dic)
```

```python
dic['name'] = "김기원"
print(dic)
```

```python
# key의 리스트를 생성하려면 : a.keys()
dic.keys()
```

```python
# key의 값들을 리스트로 생성하려면
dic.values()
```

```python
# key와 value의 쌍으로 튜플처리하려면
dic.items()
```

```python
# get을 이용해 key를 넣으면 값이 나오도록 하는 것.
dic.get("name")
```

```python
# key에 해당하는 값을 가지고 올 때 없는 key를 넣으면 에러 발생 된다??
dic["age"]   # 없는 key를 가지고오려면 오류발생 keyError
# 이것은 프로그램 실행 중 치명적인 오류로 강제 종료됨.
```

```python
dic.get("age", "값 없음")
#         k      key가 없을 때 출력문
```

```python
print(dic)
```

```python
"name" in dic  # dic 안에 name이 들어가 있냐 라는 뜻.  # in 명령어는 리스트나, 튜플, 딕셔너리 안에 값이 들어있는지...
```

```python
"age" in dic   # 이 코드로 하게 되면 key만 찾게 됨.
```

```python
"김기원" in dic
```

```python
# key를 이용해서 value -- pop (값을 꺼내와서 변수에 입력후 삭제한다.)
print(dic)
```

```python
email = dic.pop("email")
print(email)
```
