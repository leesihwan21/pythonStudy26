```python
# 튜플은 리스트와 유사하지만, 수정과 삭제가 불가능한 코드.
# 리스트는 [] 로 작성하고, 튜플은 () 작성한다.
# 1개의 값만 튜플에 넣을 때는 a= (1,) 쉼표 필수.
# b = 1,2,3,4 로 해도 튜플 처리 됨.
```

```python
# 튜플 생성은 변수에 넣은 값을 활용할 수 있다.
t1 = (1,2,'a','b')
print(t1)
```

```python
type(t1)
```

```python
# 값을 수정. 아마 오류가 나온다.
t1[3] = 8
```

```python
# 값을 삭제. 아마 오류가 나온다.
del t1[2]
```
