```python
# 숫자형(Number type) : 숫자로 이루어진 자료형으로
# 123과 같은 정수, 12.34과 같은 실수, 8진법, 16진법 등이다.
#
# 정수형(integer) : 양의정수, 음의정수, 0
a = 123
print(a) # 양의 정수
```

```python
b = -123
print(b) # 음의 정수
```

```python
# 실수형(floating point) : 소수점이 포함된 숫자
c = 1.2
print(c) # 양의 실수
```

```python
d = -3.45
print(d) # 음의 실수
```

```python
# 지수 형식 (과학기술용) 엄청 큰 실수들
e = 4.24E10 # 42,400,000,000 억
print(e)
```

```python
f = 4.24e-10 # 0.000000000424
print(f)
```

```python
# 8진수(octal) 입력 기법
# 숫자 0과 영문자 o를 쓰고 입력 한다.
g = 0o177 # 8진법 177
print(g)
```

```python
# 16진법 (hexadecimal) 0x로 입력 한다.
h = 0x8ff
print(h)
```

```python
# 사칙연산
# a+b 더하기
# a-b 빼기
# a*b 곱하기
# a**b 제곱근
4**2
```

```python
# a/b 나누기
# a//b 나머지 말고 몫 구한 값.
5//2
```

```python
5%2 # 몫 말고 나머지 구한 값.
```

```python
# 복합 연산자
# 산술연산자와 대입연산자를 합쳐 놓은 것.
i = 0
i = i + 1 # i 변수에 1을 더하고 i에 대입하라.
print(i)
```

```python
i += 1 # i = i + 1을 줄인 형태.
print(i)
```

```python
i -= 1 # i = i - 1을 줄인 형태.
print(i)
```

```python
i *= 5 # i = i * 5
print(i)
```
