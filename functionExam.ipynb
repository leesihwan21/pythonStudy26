```python
# 함수는 반복적인 코드를 위쪽에 선언하여 호출함으로써
# 프로그램을 가독성 좋고 간편하게 구현할 수 있다.


# 클래스가 없는 기능을 함수라고 부르고
# 클래스가 있는 기능을 메서드라고 한다.

# def 는 함수나 메서드를 만드는 예약어
# def 함수명(매개변수 또는 인수 또는 파라미터) :
#                                       들여쓰기 필수!
#   함수내부의 실행문들......
#   return 값 함수 종료시 호출한 곳으로 넘겨주는 값
# 들여쓰기를 끝내면 함수 종료.

# 변수 = 함수명(매개변수 == 인수 == 파라미터)
# 위에서 만든 함수를 호출하여 처리하고 return을 제공 받는다.
```

```python
# 함수를 만든다. (함수 정의)
# 맨 위에서 시작함.
def add(a, b):  # add() 함수를 만들고 인수는 a와 b를 받는다
    return a+b # 받은 a, b를 더하고 다시 돌려준다.
```

```python
# 위에서 만든 함수를 활용해보자.
c = add(5, 10) # 위에서 만든 add 함수에 5와 10을 넣고 결과를 c에 넣는다.
print(c)
```

```python
# 만약 입력값이 몇개가 될 지 모를떄는 ?????
def add_many(*args): # * 을 붙이면 몇 개의 값이 들어오든 *args ,
    # * 의 뜻은 와일드 문자로 김씨로 시작한다 ! "김*"
    # "*원" 원으로 끝나는 문자
    # "*딸기*" 글자 안에 딸기라는 글자가 있으면 추출한다. (반복된 문자를 추출)
    # args는 인수라는 뜻으로 다른 글자를 사용해도 됨.
    # 그러나 프로그래머들의 관례로 대부분 args로 사용함.
    result = "" # 결과를 보관하는 변수
    for i in args: # args는 튜플로 들어옴 (튜플은 변할수 없는값이기 때문에) 순서대로 i 값에 들어간다는 뜻.
         result = result + i # 튜플로 되어있는 값을 번갈아 더해지는 것. # for 문 안에서 변수만들면 1회용.
    return result
```

```python
# add_many 테스트
result = add_many(1,2,3,6) # 1,2,3 이라는 숫자가 args로 들어감.
print(result)
```

```python
resultText = add_many("안녕하세요","반갑습니다","파이썬 지옥입니다.") # typeerror 는 숫자 문자만 생각. # 문자열은 웬만하면 쓰지말아라.
print(resultText)
```

```python
# 함수의 리턴값은 기본 1개이지만, 더 만들수 있다.
def add_and_mul(a, b):
    return a + b, a * b   #  결과를 2개로 보낸다.
```

```python
result = add_and_mul(5,7)
print(result) # 값이 두개가 튜플로 고정된 값으로 나옴. 튜플타입으로 뤼턴된다.
```

```python
# 변수 선언에는 전역변수가 있고, 지역변수가 있다.
# 전역 변수는 함수나 if for while 문 밖에 맨 위에 작성하는 것이 좋다.
# 지역 변수는 함수나 if for while 에서 1회용으로 사용되는 것으로 안쪽에
# 선언 한다.
```

```python
g1 = 1 # 변수를 위에다가 선언 return이 없으면 안바뀌고 return을 넣어야 받아서 바뀌고 출력한다.
def vartest(g1):
    g1 = g1 + 1
    return g1 # 변수에 리턴을 넣으면 받을 수 있는 형태가 됨.
g1 = vartest(g1) # 함수 실행 후 결과를 g1 에 다시 넣음.
print(g1) # print(g1) 은 맨 위에서 선언한 것만 출력시킨다.
```

```python
a = None    # 전역 변수 a
def vartest1(a): # 지역 변수 a -> 결론은 서로 다른 a 변수이다. # 위에 있는 a하고 vatest(a) 하고는 변수가 다르다.
    a = a + 1
    # return 없음.

vartest1(3)
print(a) # a라는 변수는 로컬 변수임으로 def가 종료되면 사라짐.
```

```python
a = None # a는 글로벌 전역 변수라는 개념
def vartest2(a): # 로컬 지역 변수라는 개념
    a = a + 1
    return a # 위 코드와 변화된 내용은 return 과

a = vartest2(3) # 함수 처리 후 결과를 a에 넣음.
print(a)
```
