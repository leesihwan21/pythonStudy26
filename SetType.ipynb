```python
# 집합(set) 자료형
# 로또 주머니와 같이 중복을 허용하지 않고 순서가 없는
# 데이터들의 모임이다.
# 교집합, 합집합, 차집합 등의 집합 연산을 쉽게 처리할 수 있다.


s1 = set([1, 2, 3]) # set() 에 값을 넣으면 된다.
print(s1)
```

```python
s2 = set("hello")  # ll 영어 중복된 것을 한개만 추출, 순서도 뒤죽박죽 하게 나온다.
print(s2)
```
