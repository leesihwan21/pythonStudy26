```python
from idlelib.macosx import hideTkConsole

# 리스트란, 표 형식의 배열을 말한다.
# ids = ["kkw","lhk","ksb"]
# pws = ["1234","4321","5432","6788"]
# names = ["김기원", "임효정", "김수빈"]
# s 를 붙인다는 건 변수
# 리스트를 만들때는 [] 대괄호 로 감싸주고, ,쉼표로 구분한다.

# a = [] 빈 리스트를 생성한다. -> a = list()
# b = [1, 2, 3] 정수를 입력 할 수 있다. -> b[1] => 2
# c = [1, 2, 'kkw', 'lhk'] 와 같이 혼합하여 넣을 수 있다.
#     학년 반  이름     id
# 이렇게는 파이썬만 가능함.
# d = [1, 2, ['kkw', '1234']] -> 2차원 리스트

# d 변수에서 KKW 를 가져고오 싶으면,,,, 어떻게 해야 할까,,, d[2][0]
# f = [1, 2, ['kkw', '1234', ['파이썬', '자바']]] -> 3차원 리스트

# f 변수에서 파이썬을 가지고 오고 싶으면,,,????? f = [2][2][0]
# 리스트의 인덱싱 -> 문자열 인덱싱과 같다.
# append -> 맨 뒤에 추가를 해준다.

b = [1,2,3,['a','b','c']] # 2차원 리스트
b[3][2]

b[-1][1]

# 리스트의 슬라이싱 -> 문자열의 슬라이싱과 같다.
c = [1,2,3,4,5,6]
c[0:2] # 주의사항 마지막 숫자 -1 인덱스 값이 나온다.
c[:2] # 콜론 : 앞에 생략하면 처음값
c[2:] # 뒤에 값을 생략하면 끝까지

```

```python
# 초보자가 범하기 쉬운 리스트 연산 오류
h = [1,2,3]
# h[2]+"hi" -> 3hi
str(h[2])+"hi"
```

```python
# 리스트에 C R U D 기법을 적용.
# CREATE -> 생성, 추가, 삽입. .append()
# READ -> 전체 리스트 보기. 1개만 읽기 .index()
# UPDATE -> 자료 수정. PWs[3] = "4486"
# DELETE -> 자료 삭제. del PWs[3] .remove(3) .pop(3)
```

```python
# 리스트에 자료를 추가해보자.
IDs = ["kkw","lhj","ksb"] # 회원 id
IDs.append("mbc")
print(IDs)
```

```python
PWs = ["1234","4321","4567"]
PWs.append("8888")
print(PWs)
```

```python
# 리스트에 있는 값 가져오기
IDs.index("mbc")
```

```python
# mbc ID에 인덱스를 알아봐 패스워드의 값을 찾아옴.
PWs[3]
```

```python
# 알아온 패스워드 값을 4486으로 수정한다.
PWs[3] = "4486"
print(PWs)
```

```python
# 3번째 인덱스에 암호를 삭제한다.
del PWs[3]
print(PWs)
```

```python
# pop은 삭제되는 내용을 출력하고 삭제한다.
# PWs.pop() -> 맨 마지막 자료를 삭제.
# PWs.pop(2) -> 2번째 자리 인덱스를 삭제한다.
PWs = ["1234","4321","4567"]
PWs.pop(2)
```
