```python
# 주민번호 앞 2자리를 생년으로 추출 year
# 주민번호 3~4 자리를 생월로 추출 month
# 주민번호 5~6 자리를 생일로 추출 day
# 주민번호 8번째 1자리를 남여 추출 sex
inputSsn = "120604-2345678"
year = inputSsn[0:2]
month = inputSsn[2:4]
day = inputSsn[4:6]
print("귀하에 생년은 "+year + "년")
print("귀하에 생월은 " +month + "월")
print("귀하의 생일은 "+day + "일 입니다. ")

# 주민번호를 이용해서 남여 구분을 위한 코드
# 문자열로 인식된 숫자를 숫자로 계산시 타입변환 필요
ssnNumber = inputSsn[7]
# print(ssnNumber)
# type(ssnNumber)
ssnNumber = int(ssnNumber) # 문자가 숫자로 타입변환을 해줌. int(문자)
# type(ssnNumber)

if ssnNumber % 2 == 1 : # 숫자로 된 성별을 2로 나눈 나머지가 1인지 물어보는 것.
    sex = "남자"
else:
    sex = "여자"

print(sex)

```

```python
# 미션 주민번호 8번째 글자가 1,2,5,6이면 1900년생
# 3,4,7,8이면 2000년 생으로 만들어 보세요.
# elif 는 if 문의 첫 번째 조건이 거짓일 때 추가적인 조건을 검사하기 위해 사용되며, 여러 개의 elif 를 연속으로 여러 조건들을 순차적으로 확인할 수 있다.
# elif 뜻은 "그렇지 않고 만약 ~라면" 이라는 뜻으로 사용됨.
# 기본 구조. if 조건 1: # 조건1이 참일 때 실행. elif 조건 2: # 조건1이 거짓이고, 조건2가 참일 때 실행. else 조건 3: 위의 모든 조건들이 거짓일 때 실행.
# 예를 들어, Score = 80 점이다. if score >= 90: print("A"학점), elif score >= 70: print("B학점"), else: print("F학점")

if ssnNumber == 1 :
    yearif = "19"+year # 문자인경우 + 기호는 붙인다.
elif ssnNumber == 2 :
    yearif = "19"+year
elif ssnNumber == 5 :
    yearif = "19"+year
elif ssnNumber == 6 :
    yearif = "19"+year
elif ssnNumber == 9 :
    print("당신은 외계인 입니다. 돌아가세요.")
else:
    yearif = "20"+year

print(yearif)
```
