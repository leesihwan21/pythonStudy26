```python
# if 문은 조건문으로 판단하여 참(True), 거짓(False)로 판단하는 코드
# if 조건 :
#    조건이 참일 때 실행하는 문
# elif 조건 :
#    위 조건이 False일 때 또 비교하는 조건문
# else :
#      위 조건이 모두 해당하지 않을 때 사용.

# 학점 계산용 코드
avg = int(input("평균을 입력하세요 >>>"))
if avg >= 90 :
    grade = "A" # 90점 이상이면 A

elif avg >= 80 :
    grade = "B" # 80점 이상이면 B

elif avg >= 70 :
    grade = "C" # 70점 이상이면 C

else:
    grade = "F" # 나머지 F

print("귀하의 등급은 : " + grade + "입니다. ")

```

```python
# coffee.py
coffee = 10
while True:
    money = int(input("돈을 넣어 주세요: "))
    if money == 300:
        print("커피를 줍니다.")
        coffee = coffee -1
    elif money > 300:
        print("거스름돈 %d를 주고 커피를 줍니다." % (money -300))
        coffee = coffee -1
    else:
        print("돈을 다시 돌려주고 커피를 주지 않습니다.")
        print("남은 커피의 양은 %d개 입니다." % coffee)
    if coffee == 0:
        print("커피가 다 떨어졌습니다. 판매를 중지 합니다.")
        break
```

```python
marks = [90, 85, 77, 95, 60] # 학생들의 점수

number = 0
for mark in marks: # 90, 85, 77, 95, 60 순서대로 mark 에 각각 대입
    number = number + 1
    if mark >= 60:
        print("%d번 학생은 합격입니다." % number)
    else:
        print("%d번 학생은 불합격입니다." % number)
```

```python
# marks3.py
marks3 = [90,100,95,85,80]

for number in range(len(marks3)):
    number += 1
    if marks3[number] < 90:
        continue
    print("%d번 학생은 축하합니다. 최종합격하셨습니다." % number)
```
